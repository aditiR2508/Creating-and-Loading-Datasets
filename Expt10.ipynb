```python
#dataframe to excel
import pandas as pd

data = {
    "Roll_no": [1, 2, 3, 4, 5],
    "Gender": ["female", "female", "male", "male", "female"],
    "Department": ["EnTC", "AIML", "CSE", "EnTC", "MECH"],
    "CGPA": [9.0, 7.2, 8.8, 8.5, 7.9]
}

df = pd.DataFrame(data)

df.to_excel("student_data.xlsx", index=False)
```

```python
#accessing information of a specific column
df["Department"]

```

```python
# Accessing information of roll no. 3

roll_no_3 = df[df["Roll_no"] == 3]
print(roll_no_3)
```

```python
# To find data types of each column

df.dtypes
```

```python
# Performing mean,median and mode operation on provided data

print("Mean:", df["CGPA"].mean())
print("Median:", df["CGPA"].median())
print("Mode:", df["CGPA"].mode()[0])
```

```python
# Inserting a new column

df["Grade"] = ["A", "C+", "B+", "B+", "B"]

print(df)
```

```python
# Deleting a column

df = df.drop("Gender", axis=1)

print(df)
```
